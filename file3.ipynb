```python
table = soup.find('table',class_ = 'tblList')
print(table)
```

```python
title = table.find_all('th')
print(title)
```

```python
header = []
for i in title:
    name = i.text
    header.append(name)
```

```python
header
```

```python
import pandas as pd
df = pd.DataFrame(columns=header)
```

```python
row = table.find_all("tr")
```

```python
for i in row[1:]:
    first_td = i.find_all("td")[0].text.strip()
    data = i.find_all("td")[1:]
    row = [tr.text for tr in data]
    # print(row)
    row.insert(0,first_td)
    l = len(df)
    df.loc[l] = row
```

```python
print(df)
```
